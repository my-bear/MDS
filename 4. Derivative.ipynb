### Математический анализ. Производная

Посчитайте значение первой и второй производной функции:  

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$  

В точке x = 1.

```python
from sympy import *
import numpy as np
```

```python
# Объявляем переменую
x = symbols('x')

# Исходная функция
f = x**5 + 4*sin(2*x) + cos(3*x+3)

# Первая производная от функции f(x)
fx_1 = f.diff(x)
```

```python
fx_1
```

```python
# Считаем значение производной в точке x0
fx = lambdify(x, fx_1)

print(round(fx(1), 2))
```

```python
# Вторая производная от функции f(x)
fx_2 = fx_1.diff(x)

fx_2
```

```python
# Считаем значение производной в точке x0
fx_x = lambdify(x, fx_2)

print(round(fx_x(1), 2))
```

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
import matplotlib.pyplot as plt

#Увеличим дефолтный размер графика
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
# Возвращает равномерно расположенные числа в указанном интервале.
x = np.linspace(-5,5,250)

a = np.sin(2*x+1)**5
b = 10*np.cos(2*x+1)*np.sin(2*x+1)**4
c = -20*np.sin(2*x+1)**5+80*np.cos(2*x+1)**2*np.sin(2*x+1)**3
```

```python
fig, ax = plt.subplots()
ax.plot(x, b, 'r', label='y=f\'(x)')
ax.plot(x, c, 'g', label='y=f\'\'(x)')

# Выравниваем оси графика
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.legend(loc='upper left')
plt.show()
```
